```python
import nltk

# download once
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# for paragraph to sentences
from nltk.tokenize import sent_tokenize

corpus = "Hello world. This is NLP. You are learning fast."
print(sent_tokenize(corpus))
```

```python
# for paragraph to words
from nltk.tokenize import word_tokenize
print(word_tokenize(corpus))
```

```python
for sentence in corpus:
    print(word_tokenize(sentence))
```

```python
# stemming
# classification problem
# comment of product is a positive review or negative review
# reviews--- eating, eaten-> eat

words = ["eating","going","playing","sleeping","running","king", "moving","studies","eaten"]
```

### Porterstemmer

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem("hellowing")
```

```python
from nltk.stem import RegexpStemmer

# remove 'ing' at the end

stemmer = RegexpStemmer('ing$',min = 5)

words = ["playing", "running", "going", "sing", "king"]

for word in words:
    print(stemmer.stem(word))
```

```python
stemmer.stem('ingeating')
```

```python
from nltk.stem import SnowballStemmer

snow_stemmer = SnowballStemmer("english")

words = ["playing", "running", "happily", "studies"]

for word in words:
    print(word+"-->"+snow_stemmer.stem(word))
```

```python
snow_stemmer.stem("fairly"),snow_stemmer.stem("sportingly")
```

### Lemmatization


```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["eating","going","playing","sleeping","running","king", "moving","studies","eaten"]

for word in words:
    print(lemmatizer.lemmatize(word, pos='v'))
```
