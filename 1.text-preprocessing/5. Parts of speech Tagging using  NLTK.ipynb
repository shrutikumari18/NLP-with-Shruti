#### POS Tagging--> 
To assign grammar role (pos) in each word of the sentences.

Example:
    “I am learning NLP”

Output:
    I → Pronoun, 
    am → Verb, 
    learning → Verb, 
    NLP → Noun


Common POS Tags

Tag 	Meaning	    Example
NN  	Noun	    dog, car
VB	    Verb	    run, eat
JJ  	Adjective   big, happy
RB	    Adverb	    quickly
PRP	    Pronoun	    I, he
IN	    Preposition	in, on

```python
paragraph = """Dr. APJ Abdul Kalam was a great Indian scientist and the 11th President of India. He was born in Rameswaram in a simple family. Despite facing many difficulties, he achieved great success through hard work and determination. He is known as the “Missile Man of India” because of his important contributions to the country’s defense and space programs
Dr. Kalam was a humble and inspiring person. He always motivated young people to dream big and work hard to achieve their goals. He believed that education and dedication can help anyone succeed in life. Even today, his life teaches us to be honest, hardworking, and dedicated to our nation."""
```

```python
import nltk
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
from nltk.corpus import stopwords
# and also will have to download average perceptron tagger
nltk.download('averaged_perceptron_tagger')


for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) #converts sentences into words.
    words = [word for word in words if word not in set(stopwords.words('english'))]
    # sentences[i] =' '.join(words) 
    pos_tag = nltk.pos_tag(words)
    print(pos_tag)
```
