### Stopwords
- Stopwords are the words which removes the word which is less required in paragraph in NLP.
as - the, as, is, of, to etc.

```python
import nltk
nltk.download("stopwords")  # it will give all the stopwords with respective language.
```

```python

paragraph = "Life is a journey where we learn something new every day. Sometimes the path is easy, while at other times challenges make us stronger. A person’s true identity is shaped by their thoughts and actions. If we maintain a positive mindset and never step back from hard work, achieving any goal becomes possible. Every small success gives us the courage to move forward, and every failure teaches us something valuable, helping us improve. So always stay focused on your goals and keep moving ahead with determination."
```

```python
from nltk.corpus import stopwords
```

```python
# It will give all the stopwords according to input language.
stopwords.words("english")  
```

```python
# now we will convert the paragraph into sentences

sentences = nltk.sent_tokenize(paragraph)
type(sentences)
```

```python
sentences
```

#### Apply stopwords And Filter And then Apply Streaming.

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i]) #converts sentences into words.
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] =' '.join(words) 
```

```python
sentences
```

#### Apply stopwords And Filter And then Apply Lemmatization.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i] =' '.join(words) #convert all the words into sentences
```

```python
sentences
```

#### Apply stopwords And Filter And then Apply snowball.

```python

for i in range(len(sentences)):
    sentences[i]= sentences[i].lower()
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i] =' '.join(words) #convert all the words into sentences
```
