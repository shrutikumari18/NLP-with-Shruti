### Tokenization  
Tokenization is the process of converting any paragraph or sentences into tokens.
 Can be converted as - 
 paragraph to sentences, 
 Sentences to words
 etc.


```python
import nltk

# download once
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# for paragraph to sentences
from nltk.tokenize import sent_tokenize

corpus = "Hello world. This is NLP. You are learning fast."
print(sent_tokenize(corpus))
```

```python
# for paragraph to words
from nltk.tokenize import word_tokenize
print(word_tokenize(corpus))
```

```python
for sentence in corpus:
    print(word_tokenize(sentence))
```
