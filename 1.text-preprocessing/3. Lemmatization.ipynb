### Lemmatization
- Lemmatization is the process of converting any word into their root/base form. Used in NLP for Q&A, chatbots, text summarization.

Example:

    running → run,
    better → good,
    went → go

It is better than Stemming. as- 

Example:

    studies → studi (stemming ❌)
    studies → study (lemmatization ✅)

    but it will take much time then Stemming.

```python
import nltk
nltk.download('wordnet') # will have to download wordnet
```

To specify parts of speech is very important in this technique.

POS- 
    Noun-n,
    Verb-v,
    adjective-a,
    adverb-r

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["eating","going","playing","sleeping","running","king", "moving","studies","eaten"]

for word in words:
    print(lemmatizer.lemmatize(word, pos='v')) # very important to specify parts of speech(pos). Here v refers to verb form. 
```

```python
# We can check it also as following-
lemmatizer.lemmatize("going", pos='v')
```
